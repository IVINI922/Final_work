```python
import pandas as pd
import re
import joblib
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
from collections import Counter
```

```python
df = pd.read_csv('Petitions.csv', sep=',', header=None, names=['id', 'text', 'category'], encoding='utf-8')
df = df.dropna(subset=['text', 'category'])
df['text'] = df['text'].astype(str).str.strip()
```

```python
print(f"Загружено: {len(df)} записей")
print(f"Категории: {df['category'].value_counts().to_dict()}")
df.head()
```

```python
print("Дубликаты:", df.duplicated().sum())
```

```python
print("\n Информация:")
df.info()
```

```python
# Статистика по категориям
print("\nПо категориям:")
print(df['category'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
df['category'].value_counts().head(10).plot(kind='barh', color='steelblue')
plt.title('Топ-10 категорий обращений', fontsize=14, fontweight='bold')
plt.xlabel('Количество обращений', fontsize=12)
plt.ylabel('Категория', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Стоп-слова
ru_stopwords = set(stopwords.words('russian'))

def clean_text(text):
    """Простая очистка: нижний регистр + только буквы + стоп-слова"""
    text = text.lower()
    text = re.sub(r'[^а-яё\s]', ' ', text)  # только русские буквы
    text = ' '.join(text.split())  # лишние пробелы
    # убираем стоп-слова и короткие слова
    tokens = [w for w in text.split() if w not in ru_stopwords and len(w) > 2]
    return ' '.join(tokens)
```

```python
print("Обработка текстов...")
df['text_clean'] = df['text'].apply(clean_text)
df = df[df['text_clean'].str.len() > 10]  # убираем пустые
print(f"После очистки: {len(df)} записей")
print(f"Пример:\nИсходный: {df['text'].iloc[0][:80]}...\nОчистка: {df['text_clean'].iloc[0][:80]}...")
```

```python
for cat in df['category'].unique():
    words = ' '.join(df[df['category'] == cat]['text_clean']).split()
    top = Counter(words).most_common(5)
    word_list = ', '.join([w for w, c in top])
    print(f"\n {cat}")
    print(f"    {word_list}")
print(" Используем: text_clean → category")
print(" Исключаем: id, исходный text")
```

```python
# TF-IDF с базовыми параметрами
vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words=list(ru_stopwords),  # ← Преобразуем set в list!
    ngram_range=(1, 1),
)

X = vectorizer.fit_transform(df['text_clean'])
y = df['category']

# Разделение
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(f"Матрица: {X.shape}, Train: {X_train.shape[0]}, Test: {X_test.shape[0]}")

# Сохраняем векторизатор
joblib.dump(vectorizer, 'vectorizer.pkl')
```

```python
# Логистическая регрессия — просто и эффективно
model = LogisticRegression(max_iter=500, random_state=42)
model.fit(X_train, y_train)

# Оценка
y_pred = model.predict(X_test)
print(f" Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\n Отчёт:")
print(classification_report(y_test, y_pred))

# Сохранение
joblib.dump(model, 'model.pkl')
joblib.dump(model.classes_, 'classes.pkl')
print(" Модель сохранена: model.pkl")
```

```python
def predict_category(text, model=model, vectorizer=vectorizer):
    """Предсказать категорию + вероятности"""
    # Очистка
    clean = clean_text(text)
    # Векторизация
    vec = vectorizer.transform([clean])
    # Прогноз
    pred = model.predict(vec)[0]
    probs = dict(zip(model.classes_, model.predict_proba(vec)[0].round(3)))
    return pred, probs

# Тест
print(" Тест предсказания:")
test = "Не работает лифт, 5 этаж, парадная 2"
pred, probs = predict_category(test)
print(f"Текст: {test}")
print(f" Категория: {pred}")
print(f" Вероятности: {sorted(probs.items(), key=lambda x: -x[1])[:3]}")
```

```python
# Всё необходимое для API/приложения
artifacts = {
    'vectorizer.pkl': vectorizer,
    'model.pkl': model,
    'classes.pkl': model.classes_
}

for name, obj in artifacts.items():
    joblib.dump(obj, name)
    print(f" {name}")

print("\n Готово Модель можно использовать в приложении")
```
